This notebook reproduces results presented in the article:

[1] Skaggs, T. H., R. G. Anderson, D. L. Corwin, and D. L. Suarez (2014), Analytical steady-state solutions for water-limited cropping systems using
saline irrigation water, Water Resour. Res., 50, 9656–9674, doi:10.1002/2014WR016058.

**Abstract** Due to the diminishing availability of good quality water for irrigation, it is increasingly important that irrigation and salinity management tools be able to target submaximal crop yields and support the use of marginal quality waters. In this work, we present a steady-state irrigated systems modeling framework that accounts for reduced plant water uptake due to root zone salinity. Two explicit, closed-form analytical solutions for the root zone solute concentration profile are obtained, corresponding to two alternative functional forms of the uptake reduction function. The solutions express a general relationship between irrigation water salinity, irrigation rate, crop salt tolerance, crop transpiration, and (using standard approximations) crop yield.  Example applications are illustrated, including the calculation of irrigation requirements for obtaining targeted submaximal yields, and the generation of crop-water production functions for varying irrigation waters, irrigation rates, and crops. Model predictions are shown to be mostly consistent with existing models and available experimental data. Yet the new solutions possess advantages over available alternatives, including: (i) the solutions were derived from a complete physical-mathematical description of the system, rather than based on an ad hoc formulation; (ii) the analytical solutions are explicit and can be evaluated without iterative techniques; (iii) the solutions permit consideration of two common functional forms of salinity induced reductions in crop water uptake, rather than being tied to one particular representation; and (iv) the utilized modeling framework is compatible with leading transient-state numerical models.

### Table 1 of Ref. [1]

Functions defined in Table 1 for unstressed root water uptake density profiles:

```python
# Arguments in the following defs are:   
# L: root zone depth
# Z: scaled depth coordinate, Z = z/L

def b_4321(Z, L):
    """Unstressed uptake density, 40:30:20:10 model"""
    assert Z >= 0, L >= 0
    return (9 - 8 * Z) / (5. * L) if Z <=1 else 0

def B_4321(Z):
    """Unstressed cumulative uptake density, 40:30:20:10 model"""
    assert Z >= 0
    return (9 * Z - 4 * Z**2) / 5. if Z <=1 else 1

def invB_4321(p):
    """Inverse cumulative uptake density, 40:30:20:10 model"""
    assert 0 <= p < 1
    return (9 - math.sqrt(81 - 80 * p)) / 8.

def b_trap(Z, L):
    """Unstressed uptake density, trapezoidal model"""
    assert Z >= 0, L >= 0
    if Z <= 0.2:
        b = 5 / (3. * L)
    elif Z <= 1:
        b = (25 - 25 * Z) / (12. * L)
    else:
        b = 0
    return b

def B_trap(Z):
    """Unstressed cumulative uptake density, trapezoidal model"""
    assert Z >= 0
    if Z <= 0.2:
        B = 5 * Z / 3.
    elif Z <= 1:
        B = (50 * Z - 25 * Z * Z - 1) / 24.
    else:
        B = 1
    return B

def invB_trap(p):
    """Inverse cumulative uptake density, trapezoidal model"""
    assert 0 <= p < 1
    return 3 * p / 5. if p <= 1/3. else 1 - math.sqrt(24/25. * (1-p))

def b_exp(Z, L):
    """Unstressed uptake density, exponential model w/ delta = L/5"""
    assert Z >= 0, L >= 0
    d = L / 5.
    return math.exp(-Z * L / d) / d if Z <=1 else 0

def B_exp(Z):
    """Unstressed cumulative uptake, exponential model w/ delta = L/5"""
    assert Z >= 0
    return 1 - math.exp(-Z * 5.) if Z <=1 else 1

def invB_exp(p):
    """Inverse cumulative uptake, exponential model w/ delta = L/5"""
    assert 0 <= p < 1
    return -math.log(1 - p) / 5.

Bfuncs = {
    '4321': (B_4321, invB_4321),
    'trap': (B_trap, invB_trap),
    'exp': (B_exp, invB_exp),
}
```

### Table 3 of Ref. [1]

Functions corresponding to the solute profile analytical solutions given in Table 3:

```python
def cc0_sigmoid(Z, qr, cr, bmod):
    """ Relative concentration C/C0 using sigmoid uptake reduction function 
    Z = z/L, dimensionless depth
    qr = q0/Tp, dimensionless effective irrigation rate 
    cr = C0/C50, dimensionless effective irrigation water concentration
    bmod : string indicating the uptake desnity model: '4321','trap', 'exp'
    """
    Bmod, invB = Bfuncs[bmod]
    J = 1 - Bmod(Z) / qr
    R = cr**-3
    Q = (2 * J * R - 1) / 3
    Delta = Q**3 + R**2
    if Delta >= 0:
        S = cbrt(R + math.sqrt(Delta))
        T = cbrt(R - math.sqrt(Delta))
        relconc = S + T
    else:
        theta = np.arccos(R / math.sqrt(-Q**3))
        relconc = 2 * math.sqrt(-Q) * math.cos(theta/3)
    return relconc

def cc0_thresh(Z, qr, sc0, sct, bmod):
    """ Relative concentration C/C0 using threshold uptake reduction function 
    Z = z/L, dimensionless depth
    qr = q0/Tp, dimensionless effective irrigation rate 
    sc0 = S*C0, dimensionless irrigation parameter group
    sct = S*CT, dimensionless salt tolerance paremeter group
    bmod : string indicating the uptake desnity model: '4321','trap', 'exp'
    """
    Bmod, invB = Bfuncs[bmod]
    if qr * (1 - sc0/sct) >= 1:
        relconc = cc0_nostress(Z, qr, Bmod)
    elif sc0/sct >= 1:
        relconc = cc0_thresh_com(Z, qr, sc0, sct, Bmod)
    else:
        if Z <= invB(qr * (1 - sc0/sct)):
            relconc = cc0_nostress(Z, qr, Bmod)
        else:
            relconc = cc0_thresh_com(Z, qr, sc0, sct, Bmod)
    return relconc

def cc0_nostress(Z, qr, Bmod):
    return 1. / (1 - Bmod(Z) / qr)

def cc0_thresh_com(Z, qr, sc0, sct, Bmod):
    cms = sc0 if sc0/sct >= 1 else sct
    c0cm = min(1, sc0/sct)
    F = (1 + sct) / cms - 1
    J = 1 - Bmod(Z) / qr
    G = (1 + sct)**2 * (J - c0cm) / sc0
    A = F * math.exp(G + F)
    # NOTE: Lambert W0(A) is real for A > -1/e
    # since our A is by definition positive, W0(A) is real. 
    # special.lambertw nevertheless returns x+0j
    # need to take real(W0(A)) to drop the 0j 
    return (1. + sct) / sc0 / (1 + np.real(lambertw(A)))
```

### Relative Transpiration (Eq. 9 of Ref. [1])

Functions for predicting relative transpiration, $T_r = T/T_p$

```python
# the choice of the '4321' model in these defs is arbitrary 
# because B[Z=1] = 1 for all models by definition

def reltrans_sigmoid(qr, cr):
    """ relative transpiration rate, sigmoid reduction function
    qr: relative irrigation rate, qr = q0/Tp
    cr: relative irrigation water concentration, cr = C0/C50
    """
    return qr*(1-1./cc0_sigmoid(1., qr, cr, '4321'))
               
def reltrans_thresh(qr, sc0, sct):
    """ relative transpiration rate, threshold reduction function
    qr: relative irrigation rate, qr = q0/Tp
    sc0: relative irrigation water concentration, cr = C0/C50
    """
    return qr*(1-1./cc0_thresh(1., qr, sc0, sct, '4321'))
```

### Models of Letey et al. (1985) and Shani et al. (2007)

```python
# An implementation of models described in Refs [2] and [3].
# The root finding routines and initial values used herein
# worked for the cases I looked at, but not tested extensively 

def letey85_cropyield(AW, AWT, ECi, MHT, MHS, ETmax, Ymax):
    """Crop-water production function of Letey et al. (1985) (Ref. [2])
    see that pub for an explanation of arguments
    returns the predicted crop yield
    """
    if AW <= AWT:
        return 0
    S = Ymax / (ETmax - AWT)
    YNS = S*(AW-AWT) if AW < ETmax else Ymax
    def ECe(yd):
        return MHT + 100*yd / MHS / YNS
    def LF(yd):
        return yd/AW/S if AW < ETmax else 1 - ETmax/AW + yd/AW/S
    def func(yd):
        lf = LF(yd)
        a1 = 2.*ECe(yd)/ECi - 1./lf 
        a2 = (1./5/lf) * math.log(lf + (1-lf) * math.exp(-5))
        return a1 - a2    
    YD = newton(func, 0.01*Ymax)
    if YD < 0:
        YD = 0
    if YD > Ymax:
        YD = Ymax
    return YNS - YD

def shani07_reltrans(q0, eci, Tp, ECe50, psi_root, Ks, delta, beta, ths, 
                     thr, psi_w):
    """Relative transpiration model of Shani et al. (2007) (Ref. [3])
    see that pub for an explanation of arguments
    returns the predicted relative transpiration 
    """
    eta = delta*beta
    b = 1./10
    def psi(T):
        return psi_w/((q0-T)/Ks)**(1/eta)
    def Del_psi(T):
        return psi(T) - psi_root if psi_root < psi(T) else 0
    def Tw(T):
        return min(Tp, Del_psi(T)*(q0-T)*b)
    def theta(T):
        return thr + (ths-thr) * ((q0-T) / Ks)**(1/delta)
    def fEC(T):
        return 1./ (1 + (eci * q0*theta(T) / ECe50 / (q0 - T) / ths)**3)
    def func(T):
        return T - Tw(T) * fEC(T)
    trans = brentq(func, 0.0001, 0.9999*q0)
    return trans/Tp
    
```
